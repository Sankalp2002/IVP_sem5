```python
import PIL
from PIL import Image
loc1='horse.jpg'
loc2="saturn.png"
img1=Image.open(loc1)
img2=Image.open(loc2)

w1,h1=img1.size
w2,h2=img2.size
for i in range(h1):
    for j in range(w1):
        f=img1.getpixel((j,i))[0]
        b2=(f>>2)&1
        b3=(f>>3)&1
        b4=(f>>4)&1
        b5=(f>>5)&1
        b6=(f>>6)&1
        b7=(f>>7)&1
        r=img2.getpixel((j,i))[0]
        g=img2.getpixel((j,i))[1]
        b=img2.getpixel((j,i))[2]
        if b2==1:
            r=r|1<<0
        if b3==1:
            r=r|1<<1
        if b4==1:
            g=g|1<<0
        if b5==1:
            g=g|1<<1
        if b6==1:
            b=b|1<<0
        if b7==1:
            b=b|1<<1
        img2.putpixel((j,i),(r,g,b))
            
display(img2)
res= PIL.Image.new(img1.mode, img1.size)
for i in range(h1):
    for j in range(w1):
        r=img2.getpixel((j,i))[0]
        g=img2.getpixel((j,i))[1]
        b=img2.getpixel((j,i))[2]
        b2=(r>>0)&1
        b3=(r>>1)&1
        b4=(g>>0)&1
        b5=(g>>1)&1
        b6=(b>>0)&1
        b7=(b>>1)&1
        f=0;
        if b2==1:
            f=f|1<<2
        if b3==1:
            f=f|1<<3
        if b4==1:
            f=f|1<<4
        if b5==1:
            f=f|1<<5
        if b6==1:
            f=f|1<<6
        if b7==1:
            f=f|1<<7
        res.putpixel((j,i),(f,f,f))
display(res)
```
