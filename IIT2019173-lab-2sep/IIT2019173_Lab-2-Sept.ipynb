```python
import PIL
import numpy as np
import math
from PIL import Image
loc="input.png"
img=Image.open(loc)
display(img)
print(img.size)
```

# 3a

```python
npimg=np.array(img)
newimg = np.zeros((442,457), dtype=np.uint8)
newimg[1:441,1:456]=npimg
avg3=np.full((3,3),1/9, dtype=np.single)
#newimg[5-1:5+2,6-1:6+2].size
for i in range(1,441):
    for j in range(1,456):
        x=newimg[i-1:i+2,j-1:j+2]
        x=np.multiply(x,avg3)
        s=0;
        for k in range(3):
            for l in range(3):
                s+=x[k][l]
        if s<0: s=0
        elif s>255: s=255
        else: s=math.floor(s)
        npimg[i-1][j-1]=s
output=Image.fromarray(npimg)
display(output)
```

# 3b

```python
npimg=np.array(img)
newimg = np.zeros((444,459), dtype=np.uint8)
newimg[2:442,2:457]=npimg
avg5=np.full((5,5),1/25, dtype=np.single)
for i in range(2,442):
    for j in range(2,457):
        x=newimg[i-2:i+3,j-2:j+3]
        x=np.multiply(x,avg5)
        s=0;
        for k in range(5):
            for l in range(5):
                s+=x[k][l]
        if s<0: s=0
        elif s>255: s=255
        else: s=math.floor(s)
        npimg[i-2][j-2]=s
output=Image.fromarray(npimg)
display(output)
```

# 4a

```python
npimg=np.array(img)
newimg = np.zeros((442,457), dtype=np.uint8)
newimg[1:441,1:456]=npimg
med3=np.full((3,3),1, dtype=np.single)
for i in range(1,441):
    for j in range(1,456):
        x=newimg[i-1:i+2,j-1:j+2]
        x=np.multiply(x,med3)
        s=np.median(x)
        npimg[i-1][j-1]=s
output=Image.fromarray(npimg)
tosharp=npimg
display(output)
```

# 4b

```python
npimg=np.array(img)
newimg = np.zeros((444,459), dtype=np.uint8)
newimg[2:442,2:457]=npimg
med5=np.full((5,5),1, dtype=np.single)
for i in range(2,442):
    for j in range(2,457):
        x=newimg[i-2:i+3,j-2:j+3]
        x=np.multiply(x,med5)
        s=np.median(x)
        npimg[i-2][j-2]=s
output=Image.fromarray(npimg)
display(output)
```
